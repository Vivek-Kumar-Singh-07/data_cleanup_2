```python
#now accessing required column only in the DataFrame

newdata = pd.read_csv("Airbnb_Open_Data.csv",
                   usecols=columns_to_keep) #will only retrieve selected columns
                   

newdata.head(2)
```

```python
newdata.shape   #to ge the number of rows and columns
```

```python
newdata.rename(columns={'NAME':'name'},inplace=True)  # only one column has name in uppercase so changed it to lower to match with ohters
```

```python
#checking duplicates

newdata.duplicated().value_counts()   # There are 541 True that means we have to drop these.
```

```python
#Dropping duplciate values

newdata.drop_duplicates(inplace=True,ignore_index=True)

newdata.duplicated().value_counts()   # no true values left in the dataframe
```

```python
#Checking na vlaues

newdata.isna().sum()
```

```python
newdata.drop(columns='last review',inplace=True)  ## dropped last review as it has high null percentage
```

```python
#dropping null values

newdata.dropna(inplace=True,ignore_index=True)  
```

```python
# converting True False to 1,0
newdata['instant_bookable'].replace({True:1,False:0},inplace = True)   
```

```python
newdata['instant_bookable'].value_counts()
```

```python
#replacing $ sign with '' so we can use the columns as integer instead of string

newdata['price'] = newdata['price'].str.replace("$","")
newdata['service fee'] = newdata['service fee'].str.replace("$","")
```

```python
#removing comma 
newdata['price'] = newdata['price'].str.replace(",","")
newdata['service fee'] = newdata['service fee'].str.replace(",","")
```

```python
#removing extra spaces
newdata['price'] = newdata['price'].str.strip()
newdata['service fee'] = newdata['service fee'].str.strip()
```

```python
#Converting to integer data type
newdata['price']= newdata['price'].astype(int)
newdata['service fee'] = newdata['service fee'] .astype(int)
```
